```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, LabelEncoder
from google.colab import drive


# Aim is to predict the marks of students of the test data
```

```python
# Use the file namd 'training data' to train the model

drive.mount('/content/drive')

# Specify the correct file path in your Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/training.xlsx'

# Read the Excel file
data = pd.read_excel(file_path)
print(data)
x_train = np.array(data.iloc[:,0:8])
y_train = np.array(data.iloc[:,8]).reshape(-1,1)
# Try plotting y_train with different features
# To get an idea whether to add some features or not
# Add some features if required in x_train

# Also do label encoding for features not represented in numbers
# refer the link if not know : https://youtu.be/589nCGeWG1w?si=t2Wa7LgbUOO4RooM


def label_encoding(x_train):
    label_encoder = LabelEncoder()
    x_train[:, 0] = label_encoder.fit_transform(x_train[:, 0])
    x_train[:, 1] = label_encoder.fit_transform(x_train[:, 1])
    return x_train



x_train = label_encoding(x_train)

def feature_transform(x_train):
    poly = PolynomialFeatures(degree=2, include_bias=False)
    x_train_poly = poly.fit_transform(x_train)
    return x_train_poly

x_train = feature_transform(x_train)
print(x_train)
#   return x_train

# x_train = feature_changing(x_train)
```

```python
# def z_score(x_train):

  # ---------
    # write the code for feature scaling here
    # Your code here
  # ---------
def z_score(x_train):
    x_mean = np.mean(x_train, axis=0)
    x_std = np.std(x_train, axis=0)
    x_train_scaled = (x_train - x_mean) / x_std
    return x_train_scaled, x_std, x_mean

# Apply Z-score scaling
x_train_scaled, x_std, x_mean = z_score(x_train)

# Print the final transformed and scaled training data
print("Scaled and Transformed x_train:")
print(x_train_scaled)
print("\nStandard Deviation of each feature:")
print(x_std)
print("\nMean of each feature:")
print(x_mean)

```

```python
# def cost(x_train,y_train,w,b):

  # ---------
    # Your code here
    # Use mean square error as cost function
    # return cost
  # ---------
def cost(x_train, y_train, w, b):
    m = x_train.shape[0]
    y_pred = np.dot(x_train, w) + b
    loss = np.sum((y_pred - y_train) ** 2) / (2 * m)
    return loss

# Example initialization of weights and bias
w = np.zeros((x_train_scaled.shape[1], 1))
b = 0

# Compute cost
loss = cost(x_train_scaled, y_train, w, b)
print(f"Initial cost: {loss}")
```

```python
# def gradient_descent(x_train,y_train,w,b):


  # ---------
    # Your code here
    # Choose learning rate yourself
  # ---------
def gradient_descent(x_train, y_train, w, b):
    m = x_train.shape[0]
    learning_rate=0.01
    num_iterations=1000

    for i in range(num_iterations):
        y_pred = np.dot(x_train, w) + b
        dw = (1 / m) * np.dot(x_train.T, (y_pred - y_train))
        db = (1 / m) * np.sum(y_pred - y_train)
        w = w - learning_rate * dw
        b = b - learning_rate * db
        if i % 100 == 0:
            loss = cost(x_train, y_train, w, b)
            print(f"Iteration {i}: Cost {loss}")
    return w, b

# Example initialization of weights and bias
w = np.zeros((x_train_scaled.shape[1], 1))
b = 0
learning_rate = 0.01
num_iterations = 1000

# Run gradient descent
w, b = gradient_descent(x_train_scaled, y_train, w, b)

# Compute final cost
final_cost = cost(x_train_scaled, y_train, w, b)
print(f"Final cost: {final_cost}")


    # return w,b
```
